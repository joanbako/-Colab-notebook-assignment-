```python

import pandas as pd
import matplotlib.pyplot as plt

# ------------------ DATASET ------------------
data = {
    "Student": ["Alice","Bob","Charlie","David","Ella","Femi","Grace","Henry","Irene","John",
                "Kate","Leo","Mona","Nick","Olivia"],
    "Math": [85,78,92,70,88,95,80,67,90,76,82,89,73,91,87],
    "English": [78,82,88,74,90,85,79,80,92,81,86,77,84,89,83],
    "Science": [90,85,87,69,88,94,83,72,91,77,80,88,75,93,86]
}

df = pd.DataFrame(data)

# ------------------ HISTOGRAM ------------------
plt.figure()
plt.hist(df["Math"], bins=5)
plt.title("Histogram of Math Scores")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.show()

# ------------------ SCATTER PLOT ------------------
plt.figure()
plt.scatter(df["Math"], df["Science"])
plt.title("Scatter Plot: Math vs Science")
plt.xlabel("Math Scores")
plt.ylabel("Science Scores")
plt.show()

# ------------------ MULTIPLE LINE CHART ------------------
plt.figure()
plt.plot(df["Student"], df["Math"], marker='o', label="Math")
plt.plot(df["Student"], df["English"], marker='o', label="English")
plt.plot(df["Student"], df["Science"], marker='o', label="Science")
plt.title("Multiple Line Chart of Student Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# ------------------ SUBPLOTS ------------------
fig, axs = plt.subplots(3, 1)

axs[0].bar(df["Student"], df["Math"])
axs[0].set_title("Math Scores")

axs[1].bar(df["Student"], df["English"])
axs[1].set_title("English Scores")

axs[2].bar(df["Student"], df["Science"])
axs[2].set_title("Science Scores")

plt.tight_layout()
plt.show()

# ------------------ INTERPRETATION ------------------
print("INTERPRETATION:")
print("1. Histogram shows distribution of Math scores.")
print("2. Scatter plot shows positive relationship between Math and Science.")
print("3. Line chart shows trends across subjects.")
print("4. Subplots compare performance in each subject.")
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# ------------------ DATASET ------------------
data = {
    "Student": ["Alice","Bob","Charlie","David","Ella","Femi","Grace","Henry","Irene","John",
                "Kate","Leo","Mona","Nick","Olivia"],
    "Math": [85,78,92,70,88,95,80,67,90,76,82,89,73,91,87],
    "English": [78,82,88,74,90,85,79,80,92,81,86,77,84,89,83],
    "Science": [90,85,87,69,88,94,83,72,91,77,80,88,75,93,86]
}

df = pd.DataFrame(data)

# ------------------ HISTOGRAM ------------------
plt.figure()
plt.hist(df["Math"], bins=5)
plt.title("Histogram of Math Scores")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.show()

# ------------------ SCATTER PLOT ------------------
plt.figure()
plt.scatter(df["Math"], df["Science"])
plt.title("Scatter Plot: Math vs Science")
plt.xlabel("Math Scores")
plt.ylabel("Science Scores")
plt.show()

# ------------------ MULTIPLE LINE CHART ------------------
plt.figure()
plt.plot(df["Student"], df["Math"], marker='o', label="Math")
plt.plot(df["Student"], df["English"], marker='o', label="English")
plt.plot(df["Student"], df["Science"], marker='o', label="Science")
plt.title("Multiple Line Chart of Student Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# ------------------ SUBPLOTS ------------------
fig, axs = plt.subplots(3, 1)

# Math subplot
axs[0].bar(df["Student"], df["Math"])
axs[0].set_title("Math Scores")
axs[0].set_xlabel("Students")
axs[0].set_ylabel("Scores")

# English subplot
axs[1].bar(df["Student"], df["English"])
axs[1].set_title("English Scores")
axs[1].set_xlabel("Students")
axs[1].set_ylabel("Scores")

# Science subplot
axs[2].bar(df["Student"], df["Science"])
axs[2].set_title("Science Scores")
axs[2].set_xlabel("Students")
axs[2].set_ylabel("Scores")

plt.tight_layout()
plt.show()

# ------------------ INTERPRETATION ------------------
print("INTERPRETATION:")
print("1. Histogram shows distribution of Math scores.")
print("2. Scatter plot shows positive relationship between Math and Science.")
print("3. Line chart shows trends across subjects.")
print("4. Subplots compare performance in each subject.")
```

The histogram shows that most students scored within the mid-to-high range in Mathematics, indicating generally good performance. The scatter plot reveals a positive relationship between Math and Science scores, suggesting that students who perform well in one tend to perform well in the other. The multiple line chart highlights variations in performance across the three subjects, with some students consistently scoring high in all areas. It also shows that Science scores are slightly higher overall compared to English. From the subplots, it is clear that a few students, such as lerry and Charlie, perform strongly across all subjects, while others like Henry have comparatively lower scores. The distribution of scores suggests that the class performance is fairly balanced, with no extreme outliers. Overall, the analysis indicates a strong academic performance with slight variations among students.
