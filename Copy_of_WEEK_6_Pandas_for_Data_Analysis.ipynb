```python
import pandas as pd

# Student dataset
data = {
    "Student": ["John", "Mary", "Peter", "Jane"],
    "Score": [85, 92, 67, 74]
}

df = pd.DataFrame(data)

# Top performing student
top_student = df.loc[df["Score"].idxmax(), "Student"]

# Lowest performing student
lowest_student = df.loc[df["Score"].idxmin(), "Student"]

print("Top performing student:", top_student)
print("Lowest performing student:", lowest_student)
```
