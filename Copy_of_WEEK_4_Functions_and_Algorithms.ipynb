2. Create function check_pass_fail()

3. Apply to 5 students

4. Print results

Submit notebook (.ipynb)


```python

def compute_average(scores):
    total = 0

    for score in scores:
        total = total + score

    average = total / len(scores)

    return average


scores = [90,80,70,60,50,40]

avg = compute_average(scores)

print("Average score:", avg)
```

```python

def check_pass_fail(score):

    if score >= 50:
        return "Pass"
    else:
        return "Fail"


print(check_pass_fail(30))
print(check_pass_fail(60))
```

```python

scores = [80,60,40,30,70]

for score in scores:
    result = check_pass_fail(score)
    print(score, result)
```
