```python

# Task 2: Create line, bar, and pie charts

import pandas as pd
import matplotlib.pyplot as plt

# Dataset
data = {
    "Student": ["Alice","Bob","Charlie","David","Ella","Femi","Grace","Henry","Irene","John"],
    "Math": [85,78,92,70,88,95,80,67,90,76],
    "English": [78,82,88,74,90,85,79,80,92,81],
    "Science": [90,85,87,69,88,94,83,72,91,77]
}

df = pd.DataFrame(data)

# --------- Line Chart ---------
plt.plot(df["Student"], df["Math"], marker='o', label="Math")
plt.plot(df["Student"], df["English"], marker='o', label="English")
plt.plot(df["Student"], df["Science"], marker='o', label="Science")
plt.title("Student Scores Line Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.legend()
plt.show()

# --------- Bar Chart ---------
df.set_index("Student")[["Math","English","Science"]].plot(kind="bar")
plt.title("Student Scores Bar Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()

# --------- Pie Chart (Math Scores Example) ---------
plt.pie(df["Math"], labels=df["Student"], autopct="%1.1f%%")
plt.title("Math Scores Pie Chart")
plt.show()
```

```python

plt.plot(df["Student"], df["Math"], marker='o', label="Math")
plt.plot(df["Student"], df["English"], marker='o', label="English")
plt.plot(df["Student"], df["Science"], marker='o', label="Science")

# Add title and axis labels
plt.title("Student Scores Line Chart")  # Title
plt.xlabel("Students")                  # X-axis label
plt.ylabel("Scores")                    # Y-axis label
plt.legend()
plt.show()
```

```python
"From the line chart, Femi scored the highest in Math, while Henry scored the lowest. The bar chart shows similar trends across English and Science. The pie chart shows each student’s contribution to total Math scores."
```
